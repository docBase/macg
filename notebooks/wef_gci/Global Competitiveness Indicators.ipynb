```python
raw_data_2 = raw_data[raw_data.Attribute.isin(['Value'])].drop(['Series', 'Series unindented', 'Placement', 'Dataset', 'Attribute', 'AVG', 'LAC', 'DEVASIA', 'MENAP', 'SSA', 'CIS', 'CEE', 'ADV', 'LIC', 'LMIC', 'UMIC', 'HICO', 'HICNO', 'ASEAN', 'STAGE1', 'TRANS1TO2', 'STAGE2', 'TRANS2TO3', 'STAGE3'], axis=1).sort_values(['Edition', 'GLOBAL ID'])
```

```python
raw_data_2
```

```python
def set_codes(arr):
    codes = arr.split('.')
    while len(codes) < 4:
        codes.insert(0, '-')
    return tuple(codes)
```

```python
raw_data_2['code_1'], raw_data_2['code_2'], raw_data_2['code_3'], raw_data_2['code_4'] = \
zip(*raw_data_2['Series code'].map(set_codes))
```

```python
raw_data_2
```

```python
raw_data_3 = pd.melt(raw_data_2, id_vars=['Edition', 'code_1', 'code_2', 'code_3', 'code_4', 'GLOBAL ID'], var_name='entity')
```

```python
raw_data_3['value'] = raw_data_3['value'].convert_objects(convert_numeric=True)
raw_data_3['value'] = raw_data_3['value'].astype(float)
raw_data_3['value'] = pd.to_numeric(raw_data_3['value'])
```

```python
data = raw_data_3.pivot_table('value', ['entity', 'Edition'], ['code_1', 'code_2', 'code_3', 'code_4', 'GLOBAL ID'])
```
